```python
import numpy as np
from numpy.random import randn
import pandas as pd

# 描画のためのライブラリです。
import matplotlib.pyplot as plt
import seaborn as sns
```

ヒストグラムの説明は以下を参照してください。
https://ja.wikipedia.org/wiki/%E3%83%92%E3%82%B9%E3%83%88%E3%82%B0%E3%83%A9%E3%83%A0

```python
# 正規分布に従う乱数を作ります。
dataset1 = randn(100)

# bins=10がデフォルトです。
plt.hist(dataset1)
```

```python
# もう一つ作ります。
dataset2 = randn(80)

# 色を変えてみましょう。
# 変数に格納することでデータは表示されない。
_ = plt.hist(dataset2, color='indianred')
```

```python
# 確率密度で描画
_ = plt.hist(dataset1, density=True)
```

```python
# サンプル数が違うので、比較出来るように確率密度で描画
# 透明度を調節して重なっても見えるように

plt.hist(dataset1, density=True, alpha=0.5, bins=20)
_ = plt.hist(dataset2, density=True, alpha=0.5, bins=20, color='indianred')
```

```python
# サンプルデータを用意
# seabornはデータフレームで渡す
data = pd.DataFrame({'data1': randn(1000), 'data2': randn(1000)})

# 同時分布（結合分布）
sns.jointplot(data, x='data1', y='data2')
```

```python
# もうすこし分かり易く
sns.jointplot(data, x='data1', y='data2', kind='hex')
```

```python
dframe_wine = pd.read_csv('data3/winequality-red.csv', sep=';')
dframe_wine.head()
```

```python
def ranker(df):
    df['alc_content_rank'] = np.arange(len(df)) + 1
    return df
```

```python
# アルコール度数で並べ替えます。
dframe_alc_rank = dframe_wine.sort_values('alcohol', ascending=False)
# groupbyのあと、先ほどの関数を適用します。
dframe_alc_rank = dframe_alc_rank.groupby('quality').apply(ranker)
```

```python
dframe_alc_rank_a = ranker(dframe_wine.sort_values('alcohol', ascending=False))
dframe_alc_rank_a.head()
```

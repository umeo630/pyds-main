```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
# サンプルデータを作ります。
d_range = pd.date_range(start='2020/1/1', end='2020/1/8')
names = ['A', 'B', 'C', 'D']
temp = {'value': [], 'variable': [], 'date': []}

for name in names:
    for day in d_range:
        temp['value'].append(np.random.random())
        temp['variable'].append(name)
        temp['date'].append(day)

dframe = pd.DataFrame(temp)
```

```python
# 入れ子になったデータができます。
dframe
```

```python
# 行にdate, 列にvariable、これをvalueで埋めます。
dframe_piv = dframe.pivot(index='date',
                          columns='variable',
                          values='value')
dframe_piv
```
